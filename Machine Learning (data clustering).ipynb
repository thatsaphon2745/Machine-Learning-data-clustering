1.) การจัดกลุ่มข้อมูล (data clustering) คืออะไร 
(คำสำคัญที่ควรอธิบายได้แก่ unsupervised learning, data clustering)<br>
ตอบ การรจัดกลุ่มข้อมูล (data cluster) คือ การแบ่ง partition ซึ่งเป็นหนึ่งในอัลกอริทึมที่ใช้ในการจัดการในส่วนของการทำงานที่มีการรวมกลุ่มกัน หรือเพื่อแยกกลุ่มให้ดูง่ายและชัดเจน การจัดการกลุ่มข้อมูลเป็นหนึ่งใน method ของ Unsupervised Learning และ และในส่วนของการ Cluster จะมี Algorithm อยู่ 2 Algorithm ที่เหมาะสมกับการจำแนกกลุ่ม คือ<br> 
Unsupervised learning (การเรียนรู้แบบไม่มีผู้สอน) เป็นหนึ่งในรูปแบบของการเรียนรู้ machine learning ที่ใช้สำหรับการวิเคราะห์และจัดระเบียบข้อมูลโดยไม่ต้องมีข้อมูลป้ายกำกับ (label ) ในการ training data

2.) ตัวอย่างการนำเทคนิคการจัดกลุ่มข้อมูลไปประยุกต์ใช้งาน<br>
ตอบ 1.การจัดกลุ่มลูกค้าออนไลน์ บริษัทออนไลน์สามารถใช้การ clustering เพื่อจัดกลุ่มลูกค้าตามพฤติกรรมการช้อปปิ้งของพวกเขา เช่น, ลูกค้าที่ซื้อเสื้อผ้าและรองเท้าบ่อยอาจจัดอยู่ในกลุ่มเดียวกัน และลูกค้าที่ซื้ออุปกรณ์อิเล็กทรอนิกส์บ่อยจะอยู่ในกลุ่มอื่นๆ จากนั้น, บริษัทสามารถสร้างกลยุทธ์การตลาดที่เข้ากับแต่ละกลุ่มลูกค้า<br>
    2.การจัดกลุ่มบทความข่าว บริษัทสื่อสิ่งพิมพ์หรือเว็บไซต์ข่าวสามารถใช้การ clustering เพื่อจัดกลุ่มบทความข่าวให้อยู่ในกลุ่มที่มีเนื้อหาคล้ายคลึงกัน นี้ช่วยในการสร้างหน้าแรกข่าวที่มีหัวข้อหลากหลายสำหรับผู้อ่าน<br>
    3.การจัดกลุ่มสินค้าในร้านค้า ร้านค้าสามารถใช้การ clustering เพื่อจัดกลุ่มสินค้าในร้านค้าออนไลน์เข้ากับกลุ่มลูกค้าที่มีความสนใจเหมือนกัน โดยการจัดกลุ่มสินค้าเหล่านี้ให้ง่ายต่อการค้นหาและเป็นการวิเคราะห์ข้อมูลที่มีประโยชน์ในการวางแผนสต็อกและการตลาด<br>
    4.การจัดกลุ่มข้อมูลทางการแพทย์ ในสาขาการแพทย์ clustering สามารถช่วยในการจัดกลุ่มผู้ป่วยตามลักษณะอาการหรือโรคที่เหมือนกัน ที่จะช่วยแพทย์ในการวินิจฉัยและรักษาอย่างเหมาะสม<br>
    5.การจัดกลุ่มอักษรและคำ การ clustering อักษรหรือคำในงานการประมวลผลภาษาสามารถช่วยในการสร้างระบบค้นหาเหล่านี้จะช่วยในการสร้างระบบค้นหาเอกสารหรือคำศัพท์ที่คล้ายคลึงกัน<br>

3.) อัลกอริทึมการจัดกลุ่มข้อมูล (clustering algorithms)<br> 
ตอบ<br>
K-Means clustering คือ อัลกอริทึมการจัดกลุ่มข้อมูลที่มีจุดมุ่งหมายในการแบ่งข้อมูลเป็น K กลุ่ม (cluster) โดยที่ K คือจำนวนกลุ่มที่ผู้ใช้ต้องการ โดยอัลกอริทึมนี้ทำงานโดยเลือกจุดเริ่มต้น (centroids) สำหรับแต่ละกลุ่มและจากนั้นนำข้อมูลในแต่ละกลุ่มมาใกล้จุดเริ่มต้นที่ใกล้ที่สุด จากนั้น จุดเริ่มต้นของแต่ละกลุ่มถูกปรับให้เข้าใกล้ข้อมูลในกลุ่มของตนเอง กระบวนการนี้ทำซ้ำไปเรื่อยๆ จนกว่าจุดเริ่มต้นจะไม่เปลี่ยนและข้อมูลจะถูกจัดกลุ่มเป็น cluster นั้นเอง <br>
Agglomerative clustering (Hierarchical clustering) คือ อัลกอริทึมการจัดกลุ่มที่เริ่มต้นด้วยการมองข้อมูลแต่ละตัวเป็นกลุ่มเดี่ยว (singleton cluster) แล้วจากนั้นรวมกลุ่มที่ใกล้ที่สุดอยู่เป็น cluster ใหม่จนกว่าจะมี cluster เดียวที่ในท้ายที่สุด โดยอัลกอริทึมนี้สร้างโครงสร้างลำดับของ cluster (หรือแผนภาพต้นไม้) ซึ่งทำให้สามารถดูข้อมูลในแบบย่อยๆได้<br>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

# Load Data

```python
#ทำการ import ข้อมูลเข้ามา เป็นข้อมูล้เกี่ยวกับ รสขาติของไอศกรีม ที่ประกอบด้วย Calories,Total Fat,Trans Fat,Carbohydrates,Sugars,Protein,Sugar Alcohol
df=pd.read_csv('https://github.com/prasertcbs/basic-dataset/raw/master/BaskinRobbins.tsv', sep='\t')
df.sample(10)
```

```python
df.info()
#ทำการเช็คจำนวนของข้อมูล และ จำนวนคอลัมน์ ประเภทของแต่ละ column และค่า mising value
```

```python
df.drop(columns=['Sugar Alcohol (g)'], inplace=True)
df.head(10)
#ลบ column Sugar Alcohol ทิ้งเนื่องจากมี missing value อยู่
```

```python
df.columns
```

```python
cols=['Calories', 'Total Fat (g)', 'Trans Fat (g)', 'Carbohydrates (g)', 'Sugars (g)', 'Protein (g)']
```

# Preprocessing

```python
from sklearn.preprocessing import scale
```

```python
cols=['Calories', 'Total Fat (g)', 'Trans Fat (g)', 'Carbohydrates (g)', 'Sugars (g)', 'Protein (g)']
## z-score ปรับขนาดของข้อมูลให้อยู่ในช่วงมาตรฐาน
data = scale(df[cols])
```

# K-means

```python
from sklearn import cluster
# สร้างโมเดล K-means ชื่อ model ที่กำหนดการแบ่ง cluster = 3
model = cluster.KMeans(n_clusters=3, random_state=42)
# ให้ model ของเราทำการ train ข้อมูล
model.fit(data)
```

```python
#แสดงผลการจัดกลุ่มลำดับ cluster ให้กับข้อมูล
model.labels_
```

```python
#ทำการเอา column model.labels_ ที่เป็นตัวบอก cluster ของข้อมูลเพิ่มลงไปใน dataframe df เพื่อเปรียบเทียบ
df['cluster_kmeans']=model.labels_
df
```

```python
#ดูจำนวนของ cluster แต่ละลำดับว่ามีจำนวนข้อมูลเท่าไหร่
sns.countplot(x='cluster_kmeans', data=df);
```

```python
df
```

```python
#ดูว่าแต่ละ cluster มีช่วงของข้อมูลแต่ละอย่างเป็นอย่างไร
df.groupby('cluster_kmeans').describe().T
```

* จากค่าข้างต้นสรุปให้เห็นว่าอัลกอลิทึม K-means จะแบ่ง cluster ออกเป็น 3 cluster โดย cluster 1 จะมีค่า calories สูงที่สุดมีค่า mean = 254.64 รองลงมา คือ cluster 2 มีค่า calories mean = 164.54 และสุดท้ายคือ cluster 0 มีค่า calories mean = 95.56 ซึ่งแตกต่างกันพอสมควร

# Agglomerative clustering (Hierarchical clustering)

```python
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch # draw dendrogram
```

```python
cols=['Calories', 'Total Fat (g)', 'Trans Fat (g)', 'Carbohydrates (g)', 'Sugars (g)', 'Protein (g)']
```

```python
#data เป็นข้อมูล z-score
# สร้าง dataframe ที่สร้างจากข้อมูลที่เป็น z-score 
X=pd.DataFrame(data, columns=cols)
X.head()
```

```python
#สร้างกราฟ dendrogram (แผนภาพเส้นรวม) โดยใช้ไลบรารี Scipy เพื่อวิเคราะห์ข้อมูลและ
#แสดงการจับกลุ่มข้อมูลที่ได้จากวิธีการจัดกลุ่มแบบ linnkage (hierarchical clustering) โดยการเชื่อมข้อมูล (linkage) โดยใช้วิธีการ "ward"
fig, ax=plt.subplots(figsize=(20, 7))
dg=sch.dendrogram(sch.linkage(X, method='ward'), ax=ax, labels=df['Flavour'].values)
```

```python
# สร้างโมเดล AgglomerativeClustering ชื่อ hc ที่กำหนดการแบ่ง cluster = 3
hc=AgglomerativeClustering(n_clusters=3)
hc
```

```python
# ให้ model hc ของเราทำการ train ข้อมูล
hc.fit(X)
```

```python
#แสดงผลการจัดกลุ่มลำดับ cluster ให้กับข้อมูล
hc.labels_
```

```python
#ทำการเอา column hc.labels_ ที่เป็นตัวบอก cluster ของข้อมูลเพิ่มลงไปใน dataframe df เพื่อเปรียบเทียบ
df['cluster_hc']=hc.labels_
df
```

```python
#ดูจำนวนของ cluster แต่ละลำดับว่ามีจำนวนข้อมูลเท่าไหร่
sns.countplot(x='cluster_hc', data=df);
```

```python
#ดูว่าแต่ละ cluster มีช่วงของข้อมูลแต่ละอย่างเป็นอย่างไร
df.groupby('cluster_hc').describe().T
```

* จากค่าข้างต้นสรุปให้เห็นว่าอัลกอลิทึม Agglomerative clustering จะแบ่ง cluster ออกเป็น 3 cluster โดย cluster 0 จะมีค่า calories สูงที่สุดมีค่า mean = 254.64 รองลงมา คือ cluster 1 มีค่า calories mean = 165.62 และสุดท้ายคือ cluster 2 มีค่า calories mean = 99 ซึ่งแตกต่างกันพอสมควร

* สรุปแล้วจากการแบ่ง cluster ออกเป็น 3 cluster จากอัลกอลิทึม K-means และ Agglomerative clustering มีค่าของการแยก cluster โดยอ้างอิงจากค่า calories mean ซึ่งมีค่าใกล้เคียงกันแค่ถูกจัดลำดับเลข cluster มาไม่เหมือนกัน ส่วนในด้านของจำนวนของแต่ละ cluster ก็นับว่าใกล้เคียงกันเช่นกันแค่เลข cluster มีลำดับที่ไม่เหมือนกันนั้นเอง
